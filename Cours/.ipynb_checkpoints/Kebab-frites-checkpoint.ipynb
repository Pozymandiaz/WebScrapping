```python
#import 
import requests
from bs4 import BeautifulSoup
```

```python
# URL de la page à scraper
url = "https://www.kebab-frites.com/meilleur-kebab/paris-d54.html"

# Faire une requête GET pour récupérer le contenu de la page
response = requests.get(url)

# Vérifier si la requête s'est bien passée
if response.status_code == 200:
    # Utiliser BeautifulSoup pour analyser le contenu HTML de la page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Trouver tous les éléments de la liste des restaurants
    restaurants = soup.find_all('div', class_='col-lg-6 col-md-6 col-sm-6 col-xs-12')

    # Liste pour stocker les informations des restaurants
    restaurants_info = []

    # Parcourir chaque restaurant et extraire les informations nécessaires
    for restaurant in restaurants[:20]:  # Prendre seulement les 20 premiers restaurants
        nom = restaurant.find('h2').text.strip()
        adresse = restaurant.find('div', class_='restaurant-ville').text.strip()
        note = restaurant.find('span', class_='note').text.strip()

        # Ajouter les informations du restaurant à la liste
        restaurants_info.append({
            'Nom': nom,
            'Adresse': adresse,
            'Note moyenne': note
        })

    # Afficher les informations des restaurants
    for info in restaurants_info:
        print(info)

else:
    print("La requête a échoué avec le code :", response.status_code)
```

```python
def nettoyer_nom(nom):
    nom_nettoye = re.sub(r'#\d+\s*-\s*', '', nom)
    return nom_nettoye
```

```python
def nettoyer_adresse(adresse):
    # Utiliser une expression régulière pour extraire la rue et le numéro de l'arrondissement
    match = re.match(r'(\d+)?\s*(.*?)(\d{5})', adresse)
    if match:
        rue = match.group(2).strip()
        code_postal = match.group(3)
        return rue, code_postal
    else:
        return None, None
```

```python
df_20_meilleurs_kebab['Rue'], df_20_meilleurs_kebab['Code Postal'] = zip(*df_20_meilleurs_kebab['Adresse'].apply(nettoyer_adresse))
```

```python
# Appliquer la fonction de nettoyage sur la colonne 'Adresse' du DataFrame en utilisant .loc[]
df_20_meilleurs_kebab.loc[:, ['Rue', 'Code Postal']] = df_20_meilleurs_kebab['Adresse'].apply(nettoyer_adresse).tolist()
```
